# 1 Preliminary dataset studies

We're beginning with the UCI Human Activites Recognision dataset which has the same characteristics we will output from our hardware (accel + gyro, 6 axes) and a similar task

- Dataset description: http://archive.ics.uci.edu/ml/datasets/Human+Activity+Recognition+Using+Smartphones
- Dataset download: http://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip


```python
from urllib import request
import os

dataset_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip'
local_datadir = './datasets'
local_filename = 'UCI_HAR.zip'
local_path = os.path.join(local_datadir, local_filename)

if not os.path.exists(local_path):
    os.makedirs(os.path.dirname(local_path), exist_ok=True)
    request.urlretrieve(dataset_url, local_path)

os.listdir(local_datadir)
```

```python
from zipfile import ZipFile

with ZipFile(local_path) as archive:
    files = [file for file in archive.namelist() if not file.startswith('__MACOS')]
    archive.extractall(path=local_datadir, members=files)
```
